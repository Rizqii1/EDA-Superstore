```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# ================== 1. Memuat dan Mengeksplorasi Data ==================

```python

file_path = r'd:\Data Analyst\File excel\Sample - Superstoree.csv'

# Membaca file dengan encoding 'iso-8859-1'
data = pd.read_csv(file_path, encoding='iso-8859-1', delimiter=';', on_bad_lines='skip')
df = data.copy()

df.head()
```

```python
print('------Info Columns------ \n')
df.info()
```

```python
#Drop column Row_ID
df.drop(columns = 'Row ID', inplace = True)
df.head()
```

```python
# Convert data type object to datetime 

df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
df.info()
```

```python
#check null values in column Order ID
duplicate_count = int(df['Order ID'].duplicated().sum())

print(duplicate_count)
```

```python
#check null values in column Order ID
null_count = df['Order ID'].isna().sum()

print(null_count)
```

# ================== 2. Visualisasi Data ==================

```python
ship = ['Standard Class', 'Second Class', 'First Class','Some Day']
ship_count = df['Ship Mode'].value_counts()

plt.pie(ship_count, labels = ship, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Most Ship Mode')

plt.show()
```

```python
segment = ['Consumer', 'Corporate', 'Home Office']
segment_count = df['Segment'].value_counts()

plt.pie(segment_count, labels = segment, autopct ='%1.1f%%' ,startangle = 90, colors = sns.color_palette('Set2'))
plt.title('Segment Needs In The Local Market')
plt.show()
```

```python
revenue_city = df.groupby('City')['Sales'].sum()
city = revenue_city.head(10)
top_10 = city.sort_values(ascending = False)

plt.figure(figsize = (10,6))
top_10.plot(kind = 'bar', color='violet')
plt.xlabel('City')
plt.title ('Revenue City by Sales')

for x, y in enumerate(top_10):
    plt.text(x, y + 10, str(y), ha = 'center',va = 'bottom')
plt.show()
```

```python
product_name = df['Product Name'].value_counts()
top_7 = product_name.head(7)

plt.figure(figsize = (10,6))
top_7.plot(kind = 'bar', color = 'violet')
plt.xlabel('Product Name')
plt.title('Top 7 Product Name by Sales')

for x, y in enumerate (top_7):
    plt.text(x, y , str(y), ha = 'center', va = 'bottom')

plt.show()
```

```python
df['Year'] = df['Order Date'].dt.year
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_sales, x='Year', y='Sales', marker='o', color='green')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```
